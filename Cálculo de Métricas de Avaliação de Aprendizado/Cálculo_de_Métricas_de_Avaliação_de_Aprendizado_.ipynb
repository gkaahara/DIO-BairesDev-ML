#Neste projeto, vamos calcular as principais m√©tricas para avalia√ß√£o de modelos de classifica√ß√£o de dados, como acur√°cia, sensibilidade (recall), especificidade, precis√£o e‚ÄØF-score.

##Primeiro, vamos definir uma matriz de confus√£o de exemplo para uma classe bin√°ria:

```python
import matplotlib.pyplot as plt

# 1. Definindo a matriz de confus√£o
VP = 50  # Verdadeiros Positivos
VN = 30  # Verdadeiros Negativos
FP = 10  # Falsos Positivos
FN = 10  # Falsos Negativos
N = VP + VN + FP + FN
```

##Agora vamos implementar cada m√©trica conforme as f√≥rmulas da Tabela 1:

```python
# 2. Fun√ß√µes para c√°lculo das m√©tricas
def calcular_sensibilidade(VP, FN):
    return VP / (VP + FN) if (VP + FN) != 0 else 0

def calcular_especificidade(VN, FP):
    return VN / (FP + VN) if (FP + VN) != 0 else 0

def calcular_acuracia(VP, VN, N):
    return (VP + VN) / N if N != 0 else 0

def calcular_precisao(VP, FP):
    return VP / (VP + FP) if (VP + FP) != 0 else 0

def calcular_fscore(precisao, sensibilidade):
    return 2 * (precisao * sensibilidade) / (precisao + sensibilidade) if (precisao + sensibilidade) != 0 else 0
```

##Vamos calcular todas as m√©tricas e exibir os resultados:

```python
# 3. C√°lculo das m√©tricas
sensibilidade = calcular_sensibilidade(VP, FN)
especificidade = calcular_especificidade(VN, FP)
acuracia = calcular_acuracia(VP, VN, N)
precisao = calcular_precisao(VP, FP)
fscore = calcular_fscore(precisao, sensibilidade)

# 4. Exibi√ß√£o dos resultados em tabela
print("\nüîç RESULTADOS NUM√âRICOS:")
print(f"| {'M√©trica':<15} | {'Valor':<8} |")
print("|-----------------|----------|")
print(f"| Sensibilidade   | {sensibilidade:.2%}  |")
print(f"| Especificidade  | {especificidade:.2%}  |")
print(f"| Acur√°cia        | {acuracia:.2%}  |")
print(f"| Precis√£o        | {precisao:.2%}  |")
print(f"| F1-Score        | {fscore:.2%}  |")
```

##Vamos agora ilustrar com a matriz de confus√£o do matplotlib:

```python
# 5. Visualiza√ß√£o Gr√°fica (Matplotlib)
plt.figure(figsize=(15, 5))

# Gr√°fico 1: Matriz de Confus√£o
plt.subplot(1, 2, 1)
confusion_matrix = [[VN, FP], [FN, VP]]
plt.imshow(confusion_matrix, cmap='Blues', interpolation='nearest')
plt.colorbar()
plt.xticks([0, 1], ['Negativo', 'Positivo'])
plt.yticks([0, 1], ['Negativo', 'Positivo'])
plt.title('Matriz de Confus√£o')
for i in range(2):
    for j in range(2):
        plt.text(j, i, str(confusion_matrix[i][j]), ha='center', va='center', color='black')

# Gr√°fico 2: Compara√ß√£o de M√©tricas
plt.subplot(1, 2, 2)
metricas = ['Sensibilidade', 'Especificidade', 'Acur√°cia', 'Precis√£o', 'F1-Score']
valores = [sensibilidade, especificidade, acuracia, precisao, fscore]
bars = plt.bar(metricas, valores, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])
plt.title('Compara√ß√£o das M√©tricas')
plt.ylim(0, 1)
plt.ylabel('Score')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height, f'{height:.2%}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
